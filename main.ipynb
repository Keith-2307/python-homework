```python
# importing csv module
import csv
from pathlib import Path
csvpath = Path("data/data.csv")

# csv file name
filename = "data.csv"  

#create/initializing import lists for the CSV data
month_list = []
profit_loss_list = []

# import the csv file
with open(csvpath) as csvfile:
   # creating a csv reader object
    csvreader = csv.reader(csvfile)
 # extracting field names through first row
    fields = next(csvreader)
    
    for row in csvreader:
        month_list.append(str(row[0]))
        profit_loss_list.append(int(row[1]))

# the total number of months included in the dataset.
total_months = len(month_list)

# the net profit/loss
net_profit_loss = 0

for X in profit_loss_list:
    net_profit_loss = net_profit_loss + X

# the average profit/loss
average_monthly_change_list = []
previous_month_amount = 0

for X in range(len(profit_loss_list)):
    if X == 0:
        previous_month_amount = profit_loss_list[X]
    else:
        monthly_change = profit_loss_list[X] - previous_month_amount
        average_monthly_change_list.append(monthly_change)
        previous_month_amount = profit_loss_list[X]

#print(average_monthly_change_list)

length = len(average_monthly_change_list)
total = sum(average_monthly_change_list)
profit_loss_average = total / length

# the min and max profit/loss and corresponding month
month_greatest_increase = ''
amount_greatest_increase = 0
month_greatest_decrease = ''
amount_greatest_decrease = 0

for X in range(len(average_monthly_change_list)):
    if average_monthly_change_list[X] > amount_greatest_increase:
        amount_greatest_increase = average_monthly_change_list[X]
        month_greatest_increase = month_list[X+1]
    elif average_monthly_change_list[X] < amount_greatest_decrease:
        amount_greatest_decrease = average_monthly_change_list[X]
        month_greatest_decrease = month_list[X+1]
print()        
print(f'Financial Analysis')
print('--------------------')
print()
# The total number of months included in the dataset
print(f'The total number of months in the dataset is: {total_months}')
print()

# The net total amount of "Profit/Losses" over the entire period
net_p_l = net_profit_loss
formatted_net_p_l = format(net_p_l, ',.2f')
print(f'The Net total Profit/Loss over the entire period was: $', formatted_net_p_l, sep = '')
print()

# The average of the changes in "Profit/Losses" over the entire period
avg_amt = profit_loss_average
formatted_avg_amt = format(avg_amt, ',.2f')
print(f'The average of the changes in Profit/Loss over the entire period was: $', formatted_avg_amt, sep ='' )
print()

# The greatest increase in profits (date and amount) over the entire period
date = month_greatest_increase
amt_i = amount_greatest_increase
formatted_greatest_increase = format(amt_i, ',.2f')
print(f'The greatest increase in profits over the entire period was in {date} for: $',formatted_greatest_increase, sep = '')
print()

# The greatest decrease in losses (date and amount) over the entire period
date2 = month_greatest_decrease
amt_d = amount_greatest_decrease
formatted_greatest_decrease = format(amt_d, ',.2f')
print(f'The greatest decrease in losses over the entire period was in {date2} for: $',formatted_greatest_decrease, sep = '')
print()
print()

# Output files
Financial_Analysis = Path('data/data.csv','Financial_Analysis.txt')

#Write the output to a text file
with open('Financial_Analysis.txt','w') as file:

    file.write('Financial Analysis')
    file.write('\n')
    file.write('----------------------------')
    file.write('\n')
    file.write(f'The total number of months in the dataset is: {total_months}')
    file.write('\n')
    file.write(f'The Net total Profit/Loss over the entire period was: ${formatted_net_p_l}')
    file.write('\n')
    file.write(f'The average of the changes in Profit/Loss over the entire period was: ${formatted_avg_amt}')
    file.write('\n')
    file.write(f'The greatest increase in profits over the entire period was in {date} for: ${formatted_greatest_increase}')
    file.write('\n')
    file.write(f'The greatest decrease in losses over the entire period was in {date2} for: ${formatted_greatest_decrease}')
    file.write('\n')

#open and read the file after the appending:

with open('Financial_Analysis.txt','r') as file:
    print(file.read())
```
